# Introduction to Matplotlib

## What is Matplotlib?
**Matplotlib** is a robust, low-level plotting library for Python that serves as a core visualization utility. It provides an object-oriented API designed to generate publication-quality, 2D (and basic 3D) static, animated, or interactive figures across multiple platforms.

* **Founder:** The toolkit was originally created by **John D. Hunter**.
* **Licensing:** It is entirely **open-source** and free to use for both personal and commercial applications.
* **Submodules:** Most everyday plotting functionalities reside within its `matplotlib.pyplot` submodule, which provides a familiar, MATLAB-like interface.

---

## Language Breakdown & Compatibility
To maximize execution performance and cross-platform flexibility, the codebase incorporates several programming languages to maintain high efficiency across different operating systems:

* **Python (~93.7%):** The primary framework, driving core user utilities and API functions.
* **C++ (~4.2%):** Used for performance-critical backend tasks and underlying graphic rendering operations.
* **Objective-C & JavaScript (<1.0%):** Embedded to ensure smooth, native user-interface compatibility across different web and desktop graphical ecosystems.

### Plotting X and Y Points
The `plot()` function draws markers (points) in a diagram. By default, it connects these points with a line. 

#### Function Parameters
The function requires two sequential parameters to specify the diagram coordinates:

* **Parameter 1**: An array containing the points on the **x-axis**.
* **Parameter 2**: An array containing the points on the **y-axis**.


````python
#plotting a line 
x_points=np.array([1,24])
y_points=np.array([5,26])

#plot a line from (1,5) to (24,26)
plt.plot(x_points,y_points)

#ploting with line
plt.show()

#ploting with dots
plt.plot(x_points,y_points,'o')
plt.show()
````

````python
"""
IF the x_points are not specified it uses the default X-points e.g 0,1,2.....
"""

y_points=np.array([3,7,8,4])
plt.plot(y_points)
plt.show()
````

````python
#plotting random x and y points 
rng=np.random.default_rng()#add seed="integer for reproducibility"
x_points=rng.standard_normal(size=(1,5))
y_points=rng.standard_normal(size=(1,5))    
print ("\n","Xpoints : ","\n",x_points,"\n","Ypoints : ","\n",y_points)

"""
keyword argument marker to emphasize each point with a specified marker:
"""
plt.plot(x_points,y_points,marker='o')
plt.show()

````

### Format Strings (`fmt`)
You can use a shortcut string notation parameter, called `fmt`, to specify the marker, line style, and color. 

#### Syntax
```text
marker|line|color
```

* **`marker`**: Any valid value from the Marker Reference.
* **`line`**: The line style (see table below).
* **`color`**: The color choice.

> **Note:** If you leave out the line value in the `fmt` parameter, no line will be plotted.

#### Code Example
```python
# Plots circle markers with a dotted line in red color
plt.plot(ypoints, 'o:r')
```

#### Line Reference


| Line Syntax | Description |
| :--- | :--- |
| `'-'` | Solid line |
| `':'` | Dotted line |
| `'--'` | Dashed line |
| `'-.'` | Dashed/dotted line |

---

### Color Reference


| Color Syntax | Description |
| :--- | :--- |
| `'r'` | Red |
| `'g'` | Green |
| `'b'` | Blue |
| `'c'` | Cyan |
| `'m'` | Magenta |
| `'y'` | Yellow |
| `'k'` | Black |
| `'w'` | White |

### Marker Size
You can use the keyword argument `markersize` (or the shorter version `ms`) to set the size of the markers.

#### Code Example
```python
# Plots circle markers with a size of 20
plt.plot(ypoints, marker = 'o', ms = 20)
```

---

### Marker Color
You can use specialized keyword arguments to control the specific colors of your markers.

#### Marker Edge Color
Use `markeredgecolor` (or `mec`) to set the color of the border of the markers.

```python
# Plots markers with a red edge (border)
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
```

#### Marker Face Color
Use `markerfacecolor` (or `mfc`) to set the color of the inside of the markers.

```python
# Plots markers with a red face (inside fill)
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
```

#### Combining Both Colors
You can use both properties together to color the entire marker.

```python
# Plots markers with both a red edge and a red face
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
```
